## Problem 2


Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.

````python
# Your solution here
def reverse(matrix, word):
  if word == 'right':
    return [list(reversed(col)) for col in zip(*matrix)]
  elif word == 'left':
    return [list(col) for col in reversed(list(zip(*matrix)))]

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(reverse(matrix, 'right'))
print(reverse(matrix, 'left'))
#word = str(input())
#print(reverse(matrix, word))
````

````python
# Your solution here
class Book:
    def __init__(self, title, author, year, isbn, copies=1):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.copies = copies

    def get_info(self):
        return f'"{self.title}" от {self.author} ({self.year}), ISBN: {self.isbn}'

    def issue_copy(self):
        if self.copies > 0:
            self.copies -= 1
            return True
        return False

    def return_copy(self):
        self.copies += 1

    def __str__(self):
        return f'"{self.title}" - {self.author}'


class User:
    def __init__(self, user_name, card_number):
        self.user_name = user_name
        self.card_number = card_number
        self.borrowed_books = []  # Список объектов Book

    def borrow_book(self, book):
        self.borrowed_books.append(book)

    def return_book(self, book_isbn):
        book_to_return = None
        for book in self.borrowed_books:
            if book.isbn == book_isbn:
                book_to_return = book
                break
        if book_to_return:
            self.borrowed_books.remove(book_to_return)
            return True
        return False

    def __str__(self):
        return f'Пользователь: {self.user_name}, Билет №: {self.card_number}'


class Library:
    def __init__(self):
        # Используем словари для быстрого поиска по уникальным идентификаторам
        self.books = {}  # key: isbn, value: Book object
        self.users = {}  # key: card_number, value: User object
        self.transaction_history = []

    def add_book(self, book):
        self.books[book.isbn] = book
        print(f'Книга "{book.title}" добавлена в библиотеку.')

    def add_user(self, user):
        self.users[user.card_number] = user
        print(f'Пользователь {user.user_name} зарегистрирован.')

    def find_book(self, isbn):
        return self.books.get(isbn)

    def find_user(self, card_number):
        return self.users.get(card_number)

    def issue_book(self, card_number, isbn):
        book = self.find_book(isbn)
        user = self.find_user(card_number)

        if not user:
            print(f"Ошибка: Пользователь с билетом №{card_number} не найден.")
            return
        if not book:
            print(f"Ошибка: Книга с ISBN {isbn} не найдена.")
            return

        if book.issue_copy():  # Если есть доступные копии
            user.borrow_book(book)
            log_entry = (f"{datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')} - "
                         f"Пользователь '{user.user_name}' взял книгу '{book.title}'.")
            self.transaction_history.append(log_entry)
            print(log_entry)
        else:
            print(f"Ошибка: Все копии книги '{book.title}' уже на руках.")

    def return_book(self, card_number, isbn):
        book = self.find_book(isbn)
        user = self.find_user(card_number)

        if not user:
            print(f"Ошибка: Пользователь с билетом №{card_number} не найден.")
            return
        if not book:
            print(f"Ошибка: Книга с ISBN {isbn} не найдена.")
            return

        if user.return_book(isbn): # Если пользователь действительно брал эту книгу
            book.return_copy()
            log_entry = (f"{datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')} - "
                         f"Пользователь '{user.user_name}' вернул книгу '{book.title}'.")
            self.transaction_history.append(log_entry)
            print(log_entry)
        else:
            print(f"Ошибка: Пользователь '{user.user_name}' не брал книгу '{book.title}'.")

    def display_transaction_history(self):
        print("\n--- История транзакций ---")
        if not self.transaction_history:
            print("История пуста.")
        else:
            for entry in self.transaction_history:
                print(entry)
        print("-------------------------\n")




if __name__ == "__main__":
    # 1. Создаем экземпляр библиотеки
    my_library = Library()

    # 2. Создаем несколько книг и пользователей и добавляем их в систему
    print("--- Инициализация системы ---")
    book1 = Book("1984", "Джордж Оруэлл", 1949, "978-5-17-086940-2", 2)
    book2 = Book("Мастер и Маргарита", "Михаил Булгаков", 1967, "978-5-389-01853-2", 3)
    book3 = Book("Дюна", "Фрэнк Герберт", 1965, "978-5-17-111162-8", 1)

    my_library.add_book(book1)
    my_library.add_book(book2)
    my_library.add_book(book3)

    user1 = User("Иван Петров", "U001")
    user2 = User("Анна Сидорова", "U002")

    my_library.add_user(user1)
    my_library.add_user(user2)
    print("-----------------------------\n")

    # 3. Реализуем сценарии выдачи и возврата книг
    print("--- Сценарии работы библиотеки ---")

    # Анна берет книгу "Дюна"
    my_library.issue_book("U002", "978-5-17-111162-8")

    # Иван берет книгу "1984"
    my_library.issue_book("U001", "978-5-17-086940-2")

    # Анна пытается взять еще одну книгу "Дюна", но ее больше нет в наличии
    my_library.issue_book("U002", "978-5-17-111162-8")

    # Иван возвращает книгу "1984"
    my_library.return_book("U001", "978-5-17-086940-2")

    # Попытка вернуть книгу, которую не брали
    my_library.return_book("U002", "978-5-17-086940-2")
    print("---------------------------------\n")

    # 4. Отображаем итоговую историю транзакций
    my_library.display_transaction_history()

    # Дополнительная информация для проверки состояния
    print("--- Текущее состояние ---")
    print(f"Количество копий книги 'Дюна': {my_library.find_book('978-5-17-111162-8').copies}")
    print(f"Книги на руках у Анны: {[str(b) for b in my_library.find_user('U002').borrowed_books]}")
    print("-------------------------")
````

Explain why list `b` changes after the execution of the following code:

```python

b = a

```

> Write your answer in markdown cell after: Операция b = a копирует не сам объект списка, а ссылку на него. В результате и a и b указывают на один и тот же объект [1, 2, 3]. ## Problem 6*

Let
$$A = \sum_{i=1}^{10000} \frac{1}{i^2},\quad B=\sum_{i=10000}^{1} \frac{1}{i^2}.$$
Calculate the values of $A$ and $B$ and compare them. What do you observe? Explain why this happens. What is the best way to calculate the value of $\sum\limits_{i=1}^{10000} \dfrac{1}{i^2}$?

````python
# Your solution here
sum_A = 0.0
for i in range(1, 10001):
    sum_A += 1 / (i**2)

sum_B = 0.0
for i in range(10000, 0, -1):
    sum_B += 1 / (i**2)

print(f"Значение A: {sum_A:.17f}")
print(f"Значение B : {sum_B:.17f}")
print(f"(A - B): {sum_A - sum_B}")
````

Суммы A и B отличаются из-за ошибок округления при работе с числами с плавающей точкой. При вычислении суммы A, к уже большой накопленной сумме прибавляются очень маленькие числа. При сложении очень большого и очень маленького числа точность меньшего числа теряется, и его вклад в общую сумму становится незначительным. При вычислении суммы B, сначала складываются самые маленькие числа. Так как они имеют схожий порядок, их сумма накапливается с минимальной потерей точности.
